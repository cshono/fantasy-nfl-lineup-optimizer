```python
# Data handling
import pandas as pd
import numpy as np
```

```python
# Bokeh libraries
from bokeh.io import output_file, output_notebook
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.layouts import row, column, gridplot
from bokeh.models.widgets import Tabs, Panel
from bokeh.models import ColumnDataSource, HoverTool, CategoricalColorMapper
```

```python
# The figure will be rendered inline a my Jupyter Notebook
output_notebook()
```

```python
df = pd.read_csv('C:/Users/isabe/Documents/GitHub/CWF_Fall2020/data/fan duel/fd_salaries_2019.csv')
```

```python
df.head()
```

```python
# Removing lines with salary = 0
remove_sal = df['fd_salary'] != 0
df = df[remove_sal]
```

```python
# Removing NaN 
df = df.dropna()
```

```python
df.head()
```

```python
#Providing Data
source = ColumnDataSource(df)
```

```python
output_notebook()

hover = HoverTool(tooltips = [("First Name", "@FirstName"), ("Last Name", "@LastName")], mode="hline")

plot = figure(
    tools=[hover,"crosshair"],
    x_axis_label='Points',
    y_axis_label='Salary'
)

plot.circle(x="fd_points",
            y="fd_salary",
            color='orange',
            hover_color='green',
            size=2,
            alpha=0.5,
            source=source)

show(plot)
```
